```python
from pdfminer.high_level import extract_text
raw_text = extract_text("adventuresofsher00doylrich.pdf")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_text(raw_text)

print(f"Total chunks created: {len(chunks)}")
print(chunks[0])  # Show the first chunk
```

```python
# Install first if not already
!pip install -U langchain-community

# Imports
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Create embeddings and vector DB
embedding_model = HuggingFaceEmbeddings()
vector_db = FAISS.from_texts(chunks, embedding_model)

# Save (optional)
vector_db.save_local("sherlock_vector_store")
```

```python
query = "How does Sherlock Holmes usually solve mysteries?"
docs = vector_db.similarity_search(query, k=3)  # get top 3 relevant chunks

# Combine all retrieved chunks into a single string for Groq
context = "\n\n".join([doc.page_content for doc in docs])
print(context)
```
